### EDA

```python
ratings['user_id'].value_counts()
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns={'rating':'num_of_ratings'}, inplace=True)
```

```python
number_ratings.columns
```

```python
final_ratings = ratings_with_books.merge(number_ratings, on='title')
```

```python
final_ratings.head()
```

```python
#takes the books with at least 50 ratings
final_ratings = final_ratings[final_ratings['num_of_ratings'] >= 50]
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
#drop duplicates
final_ratings.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_ratings.shape
```

```python
final_ratings.isnull().sum()
```

### pivot table

```python
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

### CSR matrics

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

### Model Training

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237, :]
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

#### Testing model

```python
def recommend_books(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    
    recommended_books = []
    for i in range(len(suggestions[0])):
        book = book_pivot.index[suggestions[0][i]]
        if book != book_name:
            recommended_books.append(book)
    
    return recommended_books
```

```python
book_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
print(recommend_books(book_name))
```
